```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
 

csv_path =  'nnUNet_raw_data_base/nnUNet_test_data/MRI_volume_internal_test331.csv'
csv_write_path =  'nnUNet_raw_data_base/nnUNet_test_data/MRI_volume_internal_test331_to_cat_plot.csv'
df = pd.read_csv(csv_path)

df.head()
```

```python
df_init = pd.DataFrame()

row_write_list  = [ 'Right_LES','Left_LES',             
            'Left_MF',  'Right_MF' ]  

col_list = ['volume1_gd','volume1_ai', 'volume2_gd','volume2_ai', 
                'volume3_gd','volume3_ai', 'volume4_gd','volume4_ai' ]

for index, row in df.iterrows():
    ai_flag = 999
    for col in col_list:
        if 'ai' in col:
            ai_flag = 1
        if 'gd' in col:
            ai_flag = 0
        row_write = 'TBD'    
        
        if 'volume1' in col:
            row_write = 'Right_LES'
        if 'volume2' in col:
            row_write = 'Right_MF'
        if 'volume3' in col:
            row_write =  'Left_MF'
        if 'volume4' in col:
            row_write = 'Left_LES'
                
        print(row['CT_id'], row_write,  row[col], 'ai_flag', ai_flag)
        
        df = pd.DataFrame({'CT_id': row['CT_id'],  
                               'volume_class':row_write,
                            'volume' :  row[col], 
                              'ai_derived': ai_flag 
                              },index=[0])
        df_init = df_init._append(df)
        df_init.to_csv(csv_write_path )
```

```python
data_set_col = 'volume_class'

y_col = 'volume'
class_by = 'ai_derived'
```

```python

df = df_init
# sns.set(rc={"figure.figsize":(60, 25)}) #width=3, #height=4
sns.set(rc={"figure.figsize":(60, 25)}) #width=3, #height=4

plt_order = ['Left_LES', 'Right_LES', 'Left_MF',  'Right_MF']


sns.set(font_scale = 1.8)
sns.set_style("ticks")

sns.set_style( 'whitegrid',{"grid.color": ".0", "grid.linestyle": ":"})
ax = sns.catplot(x= data_set_col, y= y_col, data=df, hue= class_by,
              kind="box", order = plt_order,  height=10, aspect=1.3,showfliers = False, \
                                  boxprops=dict(alpha=.8),      
                 linewidth = 2,                  
                   palette=sns.color_palette(['#99e0b0ff','#84d1e0ff' ]),  legend = False)

ax = sns.stripplot(x= data_set_col, y= y_col,data=df, hue=class_by,order = plt_order,                    
                            jitter=True,
                            dodge=True, 
                   palette=sns.color_palette(['#1cc900ff', '#1174ecff']),                   
                            alpha=0.9,size= 3
                            )

ax.get_legend().remove()
ax.patch.set_edgecolor('black')  
# sns.despine(top=False, right=False, left=True, bottom=True)

ax.patch.set_linewidth(1) 
plt.ylim(0,550)
ax.figure.savefig('csv_and_figure/volume_331_' + class_by +'.png', dpi = 1000)
```

```python
import numpy as np
from scipy import stats

col_list = sorted(list(set(df['volume_class'].tolist())))

for col in col_list:
    print(col,':')
    df_compare = df[df['volume_class'] == col]
    ai_df = df_compare[df_compare['ai_derived']==1] 
    gd_df = df_compare[df_compare['ai_derived']==0] 

    ai_volume = ai_df['volume']
    gd_volume = gd_df['volume']
    both_volume = df_compare['volume']
    
    print(ai_volume.shape)
    ks_s, ks_p = stats.ks_2samp(ai_volume, gd_volume)

    if ks_p >0.05:
        print(col, '\t\t:ks_p_value',ks_p,' Not Significant \n')
    elif ks_p < 0.05 and ks_p >=0.01:
        print(col, '\t\t:*',ks_p,' Significant \n')
    elif ks_p < 0.01 and ks_p >=0.001:
        print(col, '\t\t:**',ks_p,' Significant \n')
    elif ks_p < 0.001 :
        print(col, '\t\t:***',ks_p,' Significant \n')            
    print()
```
